```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
h131 = pd.read_csv("homework_13.1.csv", index_col=0)
h132 = pd.read_csv("homework_13.2.csv", index_col=0)
h133 = pd.read_csv("homework_13.3.csv", index_col=0)
```

```python
h131
```

```python
first_stage_X = h131[['Z']]
first_stage_X = sm.add_constant(first_stage_X)
first_stage_y = h131['X']

first_stage_model = sm.OLS(first_stage_y, first_stage_X).fit()
h131['X_hat'] = first_stage_model.predict(first_stage_X)

second_stage_X = h131[['X_hat']]
second_stage_X = sm.add_constant(second_stage_X)
second_stage_y = h131['Y']

second_stage_model = sm.OLS(second_stage_y, second_stage_X).fit()
second_stage_model.summary()
```

```python
first_stage_X = h132[['Z2']]
first_stage_X = sm.add_constant(first_stage_X)
first_stage_y = h132['X2']

first_stage_model = sm.OLS(first_stage_y, first_stage_X).fit()
h132['X_hat'] = first_stage_model.predict(first_stage_X)

second_stage_X = h132[['X_hat']]
second_stage_X = sm.add_constant(second_stage_X)
second_stage_y = h132['Y2']

second_stage_model = sm.OLS(second_stage_y, second_stage_X).fit()
second_stage_model.summary()
```
