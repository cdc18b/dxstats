```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_csv("homework_10.1.csv").drop("Unnamed: 0", axis=1)
data
```

```python
for day in range(0, 12):
    temp_df = data[data['time'] == day]
    results = sm.OLS(temp_df['y'], sm.add_constant(temp_df[['X', 'city']])).fit()
    print(f"Day {day}:\n", results.params, "\n")
```

```python
for city in range(0, 10):
    temp_df = data[data['city'] == city]
    results = sm.OLS(temp_df['y'], sm.add_constant(temp_df[['X', 'time']])).fit()
    print(f"City {city}:\n", results.params, "\n")
```

```python
num = 100000000

X = np.clip(np.random.normal(3, 1, (num,)), 0.01, 100)

Z = np.clip(np.random.normal(3, 1, (num,)), 0.01, 100)

Y = np.exp(np.log(X + Z) + np.random.normal(0, 1, (num,)))

data2 = pd.DataFrame({'Y': Y, 'X': X, 'Z': Z})
results2 = sm.OLS(data2['Y'], sm.add_constant(data2[['X', 'Z']])).fit()
results2.params
```

```python
num = 10000 


Z = np.random.normal(0, 1, (num,)) 

X = Z + np.random.normal(0, 1, (num,)) 

Y = 1.5 * X + 2.3 * Z + np.random.normal(0, X**2, (num,)) 

data3 = pd.DataFrame({'Y': Y, 'X': X, 'Z': Z})
results3 = sm.OLS(data3['Y'], sm.add_constant(data3[['X', 'Z']])).fit()

print(results3.bse)

X_coeffs = []
for trial in range(100):
    Y = 1.5 * X + 2.3 * Z + np.random.normal(0, X**2, (num,)) 
    data3 = pd.DataFrame({'Y': Y, 'X': X, 'Z': Z})
    results3 = sm.OLS(data3['Y'], sm.add_constant(data3[['X', 'Z']])).fit()
    X_coeffs.append(results3.params['X'])
print(np.std(X_coeffs))
    

```
